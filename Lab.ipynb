```python
import os

downloads_dir = 'downloads' # Define the downloads directory for consistency
```

### Step 1: Re-generate the fake files

```python
import random
import shutil

# Clean up existing downloads directory to start fresh (optional but recommended)
if os.path.exists(downloads_dir):
    shutil.rmtree(downloads_dir)
    print(f"Cleaned up existing directory '{downloads_dir}'.")

# Define the list of extensions
extensions = ['pdf', 'docx', 'jpg', 'png', 'txt', 'xlsx']

# Create the downloads directory if it doesn't exist
if not os.path.exists(downloads_dir):
    os.makedirs(downloads_dir)
    print(f"Directory '{downloads_dir}' created.")

# Generate 100 fake files
print("Generating 100 fake files...")
for i in range(1, 101):
    ext = random.choice(extensions)
    filename = f"file_{i:03d}.{ext}"
    filepath = os.path.join(downloads_dir, filename)
    with open(filepath, 'w') as f:
        f.write(f"This is dummy content for {filename}.\n")
        f.write("It's a placeholder file created by a script.")

print("100 fake files generated successfully!")
```

### Step 2: Organize the newly generated files

```python
import os

# Ensure the downloads directory exists and list its contents
if not os.path.exists(downloads_dir):
    print(f"The directory '{downloads_dir}' does not exist. Please create files first.")
else:
    # Get all files directly within the downloads_dir, excluding subdirectories
    files_in_downloads = [f for f in os.listdir(downloads_dir) if os.path.isfile(os.path.join(downloads_dir, f))]

    if not files_in_downloads:
        print(f"No files found in '{downloads_dir}'.")
    else:
        print(f"Found {len(files_in_downloads)} files in '{downloads_dir}'. Organizing them now...")
        for filename in files_in_downloads:
            filepath = os.path.join(downloads_dir, filename)

            # Get file extension (without the dot)
            _, ext = os.path.splitext(filename)
            if ext:
                ext = ext[1:] # Remove the leading dot
            else:
                ext = 'no_extension' # Handle files without an extension

            # Create the target directory for the extension
            target_dir = os.path.join(downloads_dir, ext)
            if not os.path.exists(target_dir):
                os.makedirs(target_dir)
                print(f"Created directory: '{target_dir}'")

            # Move the file
            new_filepath = os.path.join(target_dir, filename)
            os.rename(filepath, new_filepath)
            print(f"Moved '{filename}' to '{target_dir}'")
        print("File organization complete!")
```
